```python
# 문자열 매서드

import pandas as pd
import numpy as np

ser = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["Frist ", " Second", " Third", "Fourth"])

print(ser)

ser2 = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["Frist ", " Second", " Third", "Fourth"], dtype="string")

print(ser2)
```

```python
print(ser.str.lower())
print(ser.str.upper())
print(ser.str.len())
```
