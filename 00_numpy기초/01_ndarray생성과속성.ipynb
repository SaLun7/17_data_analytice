```python
# 배열 속성

arr1 = np.array([1, 2, 3, 4])
arr1.shape
```

```python
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

arr2.shape
```

```python
arr3 = np.array([
    [[1, 2],
     [3, 4]],
    [[5, 6],
     [7, 8]]
])

arr3.shape
```

```python
arr3 = np.array([
    [[1, 2],
     [3, 4]],
    [[5, 6],
     [7, 8]]
])

print(arr3.size)
```
