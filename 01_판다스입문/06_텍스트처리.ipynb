```python
# 문자열 매서드

import pandas as pd
import numpy as np

ser = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["Frist ", " Second", " Third", "Fourth"])

print(ser)

ser2 = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["Frist ", " Second", " Third", "Fourth"], dtype="string")

print(ser2)
```

```python
print(ser.str.lower())
print(ser.str.upper())
print(ser.str.len())
```

```python
result1 = ser.str.split("_")
result1
```

```python
result2 = ser.str.split("_", expand=True)
print(result2)
```

```python
print(ser.str.split("_").str.get(0))
```

```python
ser3 = pd.Series([
    "a1", "b2", "c3", "d4a5", "e6e7", "f8"
], dtype="string")

print(ser3)
```

```python
ser3.str.extractall(r"(\d)").groupby(level=0).agg("".join)
```

```python
ser3.str.extractall(r"([a-z])").groupby(level=0).agg("".join)
```

```python
# 숫자랑 필드랑 컬럼으로 분리

ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
```
