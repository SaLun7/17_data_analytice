```python
# 데이터프레임의 불린 인덱싱 : DataFrame객체[불린 시리즈]

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
df = titanic.loc[:9,["age", "fare"]]

print(df)
```

```python
print(~(df["age"] < 20))
```

```python
print(df.loc[~(df["age"] < 20)])
```

```python
(titanic.age >= 10) & (titanic.age < 20)
```

```python
titanic[(titanic.age >= 10) & (titanic.age < 20)]
```

```python
mask1 = (titanic.age >= 10) & (titanic.age < 20)
titanic[mask1]
```

```python
# titanic["age"]
mask2 = (titanic.age < 10) & (titanic.sex == "female")
print(titanic.loc[mask2].head())
```

```python
mask3 = (titanic["age"] < 10) | (titanic["age"] >= 60)
print(titanic[mask3])
```

```python
# query() 매서드를 활용한 필터링 : DataFrame객체.query(조건 문자열)

print(df.query("age < 20"))
```

```python
df.query("age >= 20 or fare < 15")
```

```python
# isin() 매서드 활용
# isin() 매서드를 활용한 필터링 : DataFrame객체.isin(추출값의 리스트)

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

pd.set_option("display.max_columns", 10)

mask1 = titanic["embark_town"] == "Southampton"
mask2 = titanic["embark_town"] == "Queenstown"

df_boolean = titanic[mask1 | mask2]
print(df_boolean.head())
```

```python
mask3 = titanic["embark_town"].isin(["Southampton", "Queenstown"])
titanic[mask3]
```
