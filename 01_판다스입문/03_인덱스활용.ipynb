```python
import pandas as pd

exam_data = {
    "이름" : ["서준", "우현", "민아"],
    "수학" : [90, 80, 70],
    "영어" : [98, 89, 95],
    "음악" : [85, 95, 100],
    "체육" : [100, 90, 90]
}

df = pd.DataFrame(exam_data)
print(df)
```

```python
ndf = df.set_index(["이름"])
print(ndf)
```

```python
ndf2 = ndf.set_index("음악")
print(ndf2)
```

```python
ndf3 = ndf.set_index(["수학", "음악"])
print(ndf3)
```

```python
# 행인덱스 재배열

# 데이터프레임객체.reindex(인덱스배열)

dict_data = {
    "c0" : [1, 2, 3],
    "c1" : [4, 5, 6],
    "c2" : [7, 8, 9],
    "c3" : [10, 11, 12],
    "c4" : [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=["r0", "r1", "r2"])
print(df)
```

```python
new_index = ["r0", "r1", "r2", "r3", "r4"]
ndf = df.reindex(new_index)
print(ndf)
```

```python
new_index = ["r0", "r1", "r2", "r3", "r4"]
ndf2 = df.reindex(new_index, fill_value=0)
print(ndf2)
```

```python
# 행 인덱스 초기화

dict_data = {
    "c0" : [1, 2, 3],
    "c1" : [4, 5, 6],
    "c2" : [7, 8, 9],
    "c3" : [10, 11, 12],
    "c4" : [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=["r0", "r1", "r2"])
print(df)
```

```python
ndf = df.reset_index()
print(ndf)
```

```python
ndf2 = df.reset_index(names=["c00"])
print(ndf2)
```

```python
ndf3 = df.reset_index(drop=True)
print(ndf3)
```

```python
# 행 인덱스를 기준으로 데이터프레임 정렬

# 행 인덱스 기준 정렬 : DataFrame객체.sort_index()

dict_data = {
    "c0" : [1, 2, 3],
    "c1" : [4, 5, 6],
    "c2" : [7, 8, 9],
    "c3" : [10, 11, 12],
    "c4" : [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=["c0", "c1", "c2"])
print(df)
```

```python
ndf = df.sort_index(ascending = False)
print(ndf)
```

```python
import pandas as pd

dict_data = {
    "c0" : [1, 2, 3],
    "c1" : [4, 5, 6],
    "c2" : [7, 8, 9],
    "c3" : [10, 11, 12],
    "c4" : [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=["r0", "r1", "r2"])
print(df)

ndf = df.sort_index(ascending=False)
print(ndf)
```

```python
dict_data = {
    "c0" : [1, 2, 3],
    "c1" : [4, 5, 6],
    "c2" : [7, 8, 9],
    "c3" : [10, 11, 12],
    "c4" : [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=["r0", "r1", "r2"])
print(df)

ndf = df.sort_values(by="c1", ascending=False)
print()
```
