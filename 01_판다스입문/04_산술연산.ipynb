```python
# NAN값이 있는 시리즈 연산

import pandas as pd
import numpy as np

student1 = pd.Series({"국어" : np.nan, "영어" : 80, "수학" : 90})
student2 = pd.Series({"수학" : 80, "국어" : 90})

print(student1)
print(student2)
```

```python
add = student1 + student2
sub = student1 - student2
mul = student1 * student2
div = student1 / student2

print(add)
print(sub)
print(mul)
print(div)
```

```python
result = pd.DataFrame([
    add,
    sub,
    mul, 
    div
] , index=["덧셈", "뺄셈", "곱셈", "나눗셈"])

print(result)
```

```python
# 연산메서드

sr_add2 = student1.add(student2)
sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)

# print(sr_add2)
# print(sr_add)

result = pd.DataFrame([
    sr_add,
    sr_sub,
    sr_mul,
    sr_div
])

print(result)
```

```python
# 데이터프레임 vs 데이터프레임

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
df = titanic.loc[:,["age", "fare"]]
print(df.tail())
```

```python
addition = df + 10
print(addition.tail())
```

```python
substration = addition - df
print(substration.tail())
```

```python
sample1 = addition.tail()
sample2 = substration.tail().fillna(0.0)
print(sample2)
```

```python
df_add = sample1.add(sample2, fill_value=0)
df_sub = sample1.sub(sample2, fill_value=0)
df_mul = sample1.mul(sample2, fill_value=0)
df_div = sample1.div(sample2, fill_value=0)

print(df_add)
print(df_sub)
print(df_mul)
print(df_div)
```
